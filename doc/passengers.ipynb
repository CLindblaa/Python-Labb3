```python
import numpy as np

def compute_passenger_stats(data):
    # Beräkna lägsta, högsta och genomsnittliga antalet passagerare för varje månad
    passenger_stats = {}
    for month, values in data.items():
        min_passengers = min(values)
        max_passengers = max(values)
        avg_passengers = np.mean(values)
        passenger_stats[month] = {
            'Lägsta': min_passengers,
            'Högsta': max_passengers,
            'Genomsnitt': avg_passengers
        }
    return passenger_stats

print(compute_passenger_stats(passenger_data))
```

```python
average_passengers_per_month = average_passengers_per_year / 12

print(average_passengers_per_month)
```

```python
def classify_seasons(passenger_stats, average_passengers_per_month):
    # Skapa en dictionary för att lagra klassificeringen av varje månad
    season_classification = {}
    
    # Loopa igenom varje månad och dess statistik
    for month, stats in passenger_stats.items():
        average_passengers = stats['Genomsnitt']
        
        # Definiera gränserna för lågsäsong, medelsäsong och högsäsong
        lower_bound = average_passengers_per_month * 0.9
        upper_bound = average_passengers_per_month * 1.1

        # Klassificera säsongen baserat på genomsnittet av passagerarna
        if average_passengers < lower_bound:
            season = 'Lågsäsong'
        elif lower_bound <= average_passengers <= upper_bound:
            season = 'Medelsäsong'
        else:
            season = 'Högsäsong'

        season_classification[month] = season

    return season_classification

# Använd resultatet från compute_passenger_stats-funktionen
passenger_stats = compute_passenger_stats(passenger_data)

# Klassificera säsongerna
seasons = classify_seasons(passenger_stats, average_passengers_per_month)
print(seasons)

```

```python
import matplotlib.pyplot as plt

# Extrahera data för Januari från passenger_data
january_data = [value / 1000 for value in passenger_data['Januari']]  # Konvertera till tusentals passagerare

# Skapa en lista med årtalen för x-axeln
years = list(range(2023, 2010, -1))  # Omvänd ordning för att börja med det senaste året

# Skapa diagrammet
plt.figure(figsize=(10, 6))
plt.plot(years, january_data, marker='o', linestyle='-')
plt.title('Passagerarantal för Januari över åren')
plt.xlabel('År')
plt.ylabel('Passagerarantal (tusen)')
plt.grid(True)
plt.xticks(years)
plt.ylim(bottom=0) 
plt.tight_layout()
plt.show()
```

## Genomsnittligt passagerarantal per månad över åren

Detta stapeldiagram ovan visar det genomsnittliga antalet passagerare per månad över en flerårsperiod. 
Varje stapel representerar genomsnittet av passagerarantalet för varje månad från januari till december. Diagrammet ger en överblick över månatliga variationer i passagerarantalet och gör det möjligt att identifiera trender eller mönster över åren.





**Låt oss gå vidare och titta på säsongsfördelning här nedan**

```python
colors = {'Lågsäsong': '#faff94', 'Medelsäsong': '#fcc856', 'Högsäsong': '#f76300'}
months = ['Januari', 'Februari', 'Mars', 'April', 'Maj', 'Juni', 'Juli', 'Augusti', 'September', 'Oktober', 'November', 'December']

# Skapa data för cirkeldiagrammet
sizes = [1] * len(classed_seasons)
labels = list(classed_seasons.keys())
pie_colors = [colors[season] for season in classed_seasons.values()]

# Sortera månaderna så att januari kommer först
sorted_labels = sorted(labels, key=lambda month: months.index(month))

# Skapa cirkeldiagram
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(sizes, labels=sorted_labels[::-1], colors=pie_colors[::-1], startangle=90)
plt.title('Säsongsfördelning över året baserat på passagerarantal')
ax.axis('equal')

# Lägg till en legend för att visa vilken färg som representerar lågsäsong och högsäsong
legend_pie_colors = [colors[season] for season in colors.keys()]  # Lista med färger för legenden
legend_patches = [plt.Rectangle((0,0),1,1,fc=color, edgecolor='none') for color in legend_pie_colors]
plt.legend(legend_patches, colors.keys(), loc='best')

# Lägg till linjer runt varje pajdel för att göra dem tydligare
for wedge in ax.patches:
    wedge.set_linewidth(1)
    wedge.set_edgecolor('white')

plt.show()
```

## Säsongsfördelning över året baserat på passagerarantal
Detta cirkeldiagram visar fördelningen av lågsäsong, medelsäsong och högsäsong över året baserat på det genomsnittliga passagerarantalet varje månad.

* **Lågsäsong**: 
    Månader med ett genomsnittligt passagerarantal som ligger klart under det genomsnittliga månatliga passagerarantalet.

* **Medelsäsong**: 
    Månader där det genomsnittliga passagerarantalet är nära det genomsnittliga månatliga passagerarantalet.

* **Högsäsong**: 
    Månader med ett genomsnittligt passagerarantal som ligger klart över det genomsnittliga månatliga passagerarantalet.
